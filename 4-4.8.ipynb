### 연속 오차

```python
#가우시안 프로세스 회귀 모델 -> 불확실한 연속 측청치 가진 데이터에 유연한 비모수 함수 적합시킴
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
```

```python
import numpy as np
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C

# 모델 함수 정의
model = lambda x: np.sin(x)
xdata = np.array([1, 3, 5, 6, 8]).reshape(-1, 1)
ydata = model(xdata).ravel()

# 커널 정의 (RBF 커널 사용)
kernel = C(1.0, (1e-3, 1e3)) * RBF(1.0, (1e-2, 1e2))

# 가우시안 프로세스 회귀 모델 정의
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10, alpha=1e-10)

# 모델 학습
gp.fit(xdata, ydata)

# 예측 데이터
xfit = np.linspace(0, 10, 1000).reshape(-1, 1)
yfit, sigma = gp.predict(xfit, return_std=True)

# 예측 결과 시각화
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(xdata, ydata, 'ro', label='Training data')
plt.plot(xfit, yfit, 'b-', label='Prediction')
plt.fill_between(
    xfit.ravel(),
    yfit - 2 * sigma,
    yfit + 2 * sigma,
    color='blue',
    alpha=0.2,
    label='95% confidence interval',
)
plt.title("Gaussian Process Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
## plt.fill_between 함수를 이용해 연속적인 오차를 옅은 색으로 시각화할 수 있음
```

```python
fig, ax = plt.subplots()

lines=[]
styles=['-','--','-.',':']
x=np.linspace(0,10,1000)
for i in range(4):
    lines+=ax.plot(x,np.sin(x-i*np.pi/2), styles[i], color='black')
ax.axis('equal')

#첫번째 범례의 라인과 레이블 지정
ax.legend(lines[:2],['line A', 'line B'],
         loc= 'upper right', frameon=False)

#두번째 범례를 만들고 아티스트 추가
from matplotlib.legend import Legend
leg= Legend(ax, lines[2:], ['line C', 'line D'], loc = 'lower right', frameon= False)
ax.add_artist(leg)
```

### 색상 지도 선택하기

#### 정성적 색상 지도 : 특정 순서 없이 색을 혼합한다. (rainbow 또는 jet)

```python
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.pyplot as plt
import numpy as np

def grayscale_map(cmap):
    # 컬러맵 가져오기
    cmap = plt.cm.get_cmap(cmap)
    colors = cmap(np.linspace(0, 1, cmap.N))  # cmap에서 색상 배열 가져오기

    # 회색조 밝기 변환
    RGB_weight = [0.299, 0.587, 0.144]
    luminance = np.sqrt(np.dot(colors[:, :3]**2, RGB_weight))
    colors[:, :3] = luminance[:, np.newaxis]  # 회색조 적용

    return LinearSegmentedColormap.from_list(cmap.name + "_gray", colors, cmap.N)

def view_colormap(cmap_name):
    # 원본 및 회색조 컬러맵 가져오기
    original_cmap = plt.cm.get_cmap(cmap_name)
    grayscale_cmap = grayscale_map(cmap_name)

    # 색상 배열
    original_colors = original_cmap(np.linspace(0, 1, original_cmap.N))
    grayscale_colors = grayscale_cmap(np.linspace(0, 1, grayscale_cmap.N))

    # 시각화
    fig, ax = plt.subplots(2, figsize=(6, 2), subplot_kw=dict(xticks=[], yticks=[]))
    ax[0].imshow([original_colors], extent=[0, 10, 0, 1])
    ax[0].set_title("Original Colormap")
    ax[1].imshow([grayscale_colors], extent=[0, 10, 0, 1])
    ax[1].set_title("Grayscale Colormap")
    plt.tight_layout()
    plt.show()

# 테스트 실행
view_colormap('jet')
```

#### 순차적 색상지도: <br>하나의 연속적인 순서
를 가지는 색상으로 구성돼 있다. ex.binary viridis


```python
view_colormap('viridis') #회색조로 인쇄해도 잘 나오게끔
```

#### 발산적 색상지도: <br> 일반적으로 평균을 기준으로 양수 및 음수 편차를 나타내는 두 가지 구별된 색상을 포함한다. (ex.RdBu 또는 Pu0r)

```python
view_colormap('RdBu')
```

### 예제: 손으로 쓴 숫자

```python
#0~5까지 숫자의 이미지를 적재하고 그 중 몇 가지를 시각화
from sklearn.datasets import load_digits
digits=load_digits(n_class=6)

fig, ax = plt.subplots(8,8,figsize=(6,6))
for i, axi in enumerate(ax.flat):
    axi.imshow(digits.images[i], cmap='binary')
    axi.set(xticks=[], yticks=[])
```

--> 차원 축소

```python
#IsoMap을 사용해 숫자를 2차원에 사영
from sklearn.manifold import Isomap
iso=Isomap(n_components=2)
projection =iso.fit_transform(digits.data)
```

```python
#결과 플롯 그리기
plt.scatter(projection[:,0],projection[:,1],lw=0.1, c=digits.target,cmap=plt.cm.get_cmap('RdBu',40))
plt.colorbar(ticks=range(6), label='digit value')
plt.clim(-0.5,5.5)
```
